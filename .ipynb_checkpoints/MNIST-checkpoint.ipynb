```python
def disturb_image(raw_data,t=0.2):
    data=pd.DataFrame(raw_data.copy())
    v=np.random.uniform(size=(28,28))
    mask=(v<=t)
    data[mask]=75
    return data,mask
    
```

```python
def show_disturb_images(i=10,t=0.2,save=False,name=None):
    j=0
    f,axes=plt.subplots(2,i,sharey=True,figsize=(i+2,4),sharex=True)
    for l in range(i):
        axes[0][l].imshow(train_X[j],cmap=plt.get_cmap('gray'))
        data,mask=disturb_image(train_X[j],t)
        axes[1][l].imshow(data,cmap=plt.get_cmap('gray'))
        axes[0][l].set_title(" ".join(["Label :",str(train_y[j])]))
        j+=1
    if save:
        directory = os.getcwd()
        tmp_fig_path="\ ".join([directory,name])
        fig_path=tmp_fig_path.replace(" ","")
        plt.savefig(fig_path)
    plt.tight_layout()
    plt.show()
    
```
